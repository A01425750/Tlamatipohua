## 1a. aproximación: grado de presión por RHA 2003-2022
geojson https://sinav30.conagua.gob.mx:8080/SINA/?opcion=base

```python
anos = np.arange(2003,2022+1)
for ano in anos:
    f = f"../datos/Grado_de_Presion_{ano}.xlsx"
    if ano == 2003:
        gp = pd.read_excel(f,usecols=["Región Hidrológico-Administrativa","Grado de Presión (%)"])
        gp.columns = ["RHA",ano]
    else:
        gp_ano = pd.read_excel(f,usecols=["Grado de Presión (%)"])
        gp_ano.columns = [ano]
        gp[ano] = gp_ano[ano]
gp["RHA"] = gp["RHA"].str.replace(r"^\s*[IVXLCDM]+\s*[-.]?\s*", "", regex=True)
name_mapping = {
    "Península de Baja California": "Península De Baja California",
    'Cuencas Centrales del Norte': 'Cuencas Centrales Del Norte',
    "Península de Yucatán": "Península De Yucatán",
    'Aguas del Valle de México': 'Valle de México'
}
gp['RHA'] = gp['RHA'].replace(name_mapping)
gp
```

```python
gp_largo = pd.melt(gp,id_vars=["RHA"],var_name="Año",value_name="Grado de presión (%)")
gp_largo["Año"] = gp_largo["Año"].astype(int)
geojson = "../datos/Región Hidrológico-Administrativa 2023.json"
fig = px.choropleth(gp_largo,geojson=geojson,locations="RHA",featureidkey="properties.rha",color="Grado de presión (%)",animation_frame="Año",color_continuous_scale="Viridis")
fig.update_geos(fitbounds="locations",visible=False)
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

```python
px.line(gp_largo,x="Año",y="Grado de presión (%)",color="RHA")
```

```python
years = np.arange(2020, 2023 + 1)
for year in years:
    f = f"../datos/repda_estatal{year}.xlsx"

    if year == 2020:
        repda = pd.read_excel(f, usecols=["Estado", "Total por estado"])
        repda.columns = ["Estado", year]
        repda[year] = (
            repda[year]
            .astype(str)
            .str.replace(",", "", regex=False)
            .str.replace(r"[^\d.\-]", "", regex=True)
        )
        repda[year] = pd.to_numeric(repda[year], errors="coerce")
    else:
        repda_year = pd.read_excel(f, usecols=["Total por estado"])
        repda_year.columns = [year]
        repda_year[year] = (
            repda_year[year]
            .astype(str)
            .str.replace(",", "", regex=False)
            .str.replace(r"[^\d.\-]", "", regex=True)
        )
        repda_year[year] = pd.to_numeric(repda_year[year], errors="coerce")
        repda[year] = repda_year[year]

repda
```
